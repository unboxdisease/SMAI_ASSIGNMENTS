#  Logistic Regression Using Gradient Descent 

### Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X, y =  make_blobs(n_samples=100, centers=[[2,4],[4,2]], random_state=20)
```

```python
#Visualize dataset 
plt.plot(X[:,0][y==0],X[:,1][y==0],'o',color='red')
plt.plot(X[:,0][y==1],X[:,1][y==1],'^',color='blue')
```

```python
def sigmoid(v):
  return 1/(1+np.exp(-v))

def J(w,x,y):
  return np.sum((sigmoid(np.matmul(w,x.T))-y)**2)/2

def loss(w,x,y):
  h = sigmoid(np.matmul(w,x.T))
  return np.mean(-y*np.log(h)-(1-y)*np.log(1-h))

def grad_descent(w,x,y):
  return np.dot(sigmoid(np.matmul(w,x.T))-y,x)/x.shape[0]

def logistic_regression(X,y,learning_rate=0.0001,max_iter=1500,tolerance=0.0000001):
  x = np.zeros((X.shape[0],X.shape[1]+1),dtype=X.dtype)
  x[:,:X.shape[1]] = X
  x[:,X.shape[1]] = 1
  w = np.zeros(x.shape[1])
  prev_w = w.copy()
  for _ in range(max_iter):
    print(
f"""
Iteration No.:{_+1}
Accuracy is {np.sum((predict(w,X)==y).astype('int'))/y.shape[0]}
"""
    )
    dw = grad_descent(w,x,y)
    print(dw)
    w -= learning_rate * dw
    if np.sqrt(np.mean((w-prev_w)**2)) < tolerance:
      break
    prev_w = w.copy()
    print('')
  return w

def predict(w,X):
  x = np.zeros((X.shape[0],X.shape[1]+1),dtype=X.dtype)
  x[:,:X.shape[1]] = X
  x[:,X.shape[1]] = 1
  return (sigmoid(np.matmul(w,x.T))>=0.5).astype('int')
```

```python
X, y =  make_blobs(n_samples=100, centers=[[2,4],[4,2]], random_state=20)
w = logistic_regression(X,y,learning_rate=0.002,max_iter=10000)
w, np.sum((predict(w,X)==y).astype('int'))/y.shape[0]
```

```python
plt.figure(figsize=(10,10))
plt.plot(X[:,0][y==0],X[:,1][y==0],'o',color='red',label='0')
plt.plot(X[:,0][y==1],X[:,1][y==1],'^',color='blue',label='1')
x = np.linspace(np.amin(X[:,0]),np.amax(X[:,0]),100)
m = -w[0]/w[1]
c = -w[2]/w[1]
plt.plot(x,x*m+c,label='Decision Boundary')
plt.legend()
plt.title('Logistic Regression')
plt.show()
```
