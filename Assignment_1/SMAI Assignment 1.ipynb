## Q1:
 **PMF with finite range**: 
*Binomial distribution*   

Consider a random variable $X$~$B(p)$ for a real number $p$ such that $0<p<1$

So, the PMF is $P(X=x)=p$ if $x=0$ and, $1-p$ if $x=1$

The range of the PMF is $R_X=\{0,1\}$ and is finite.  
  
$0<p<1\Rightarrow0<1-p<1$ and,  
$\begin{aligned}
    \sum_{x\in R_X}P(X=x)&=P(X=0)+P(X=1)\\
    &=p+1-p\\
    &=1
    \end{aligned}$  
Hence, this is a valid PMF.

 **PMF with infinite range** : *A random variable $X$ which denotes the number of coin flips till the first heads is observed.*

So, the PMF is $P(X=x)=\frac{1}{2^x}$ where $x\in N$  
The range of the PMF is $R_X=\ N$ and it is infinite.  
$\forall x\in\ N,0<\frac{1}{2^x}<1$ and,    
$\begin{aligned}
    \sum_{x\in R_X}P(X=x)&=\sum_{x\in\ N}\frac{1}{2^x}\\
    &=\frac{1}{2}\sum_{x\in\mathbb{W}}\frac{1}{2^x}\\
    &=\frac{1}{2}\times\frac{1}{1-\frac{1}{2}}\\
    &=1
    \end{aligned}$   
Hence, this is a valid PMF.

## Q2:
**Uniform density distribution PDF is given by** 
  
$U(a,b) = \frac{1}{b-a}$ if $a\leq x\leq b$ and, $0$ otherwise.  


For a random variable $X,Var(X)=E[X^2]-(E[X])^2$  
$\begin{aligned}
E[U]&=\int_{a}^{b}\frac{x}{b-a}dx\\
&=\frac{a+b}{2}
\end{aligned}$  
and,    
$\begin{aligned}
E[U^2]&=\int_a^b\frac{x^2}{b-a}dx\\
&=\frac{b^2+ab+a^2}{3}
\end{aligned}$  
Thus,  
$\begin{aligned}
Var(U)=\sigma^2&=E[U^2]-(E[U])^2\\
&=\frac{b^2+ab+a^2}{3}-\left(\frac{a+b}{2} \right)^2\\
&=\frac{b^2-2ab+a^2}{12}\\
&=\frac{(b-a)^2}{12}
\end{aligned}$  
**Hence, the given equation denotes the variance of uniform density distribution.**

## Q4 :
Let $X$ be a discrete random variable and $R_X$ the range of values taken by it,  
PMF of $X$ is :   
$P_X(x)=P(X=x)$   
variance $\sigma^2=E[(X-E[X])^2]$  
  
$\begin{aligned}  
E[(X-E[X])^2]&=\sum_{x\in R_X}P_X(x)(x-\mu)^2\\
&=\sum_{x\in R_X}P_X(x)(x^2-2x\mu+\mu^2)\\
&=\sum_{x\in R_X}P_X(x)(x^2) - \sum_{x\in R_X}P_X(x)(2x\mu) + \sum_{x\in R_X}P_X(x)(\mu^2)\\
&=E[X^2]+\mu^2-2\mu\sum_{x\in R_X}P_X(x)(x)\\
&=E[X^2]+\mu^2-2\mu E[X]\\
&=E[X^2]-\mu^2=E[X^2]-(E[X])^2
\end{aligned}$  
  
$\therefore\sigma^2=E[X^2]-(E[X])^2$  
This equation is valid for discrete random variables too.

## Q5:  
For a normal random variable $X$~$N(\mu,\sigma^2)$, the PDF is $f_X(x)=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$  
Now, $E[X]=\int_{-\infty}^{\infty}xf_X(x)dx$, $E[X^2]=\int_{-\infty}^{\infty}x^2f_X(x)dx$ and $Var(X)=E[(X-E[X])^2]$  
  
$\begin{aligned}
E[X]&=\int_{-\infty}^\infty xf_X(x)dx\\
&=\int_{-\infty}^\infty\frac{x}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}dx
\end{aligned}$   
By substituting $y=x-\mu$, and $y=x-\mu\Rightarrow dy=dx$.   
$\begin{aligned}
E[X]&=\int_{-\infty}^\infty\frac{x}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}dx\\
&=\int_{-\infty}^\infty\frac{y+\mu}{\sigma\sqrt{2\pi}}e^{-\frac{y^2}{2\sigma^2}}dy\\
&=\int_{-\infty}^\infty\frac{y}{\sigma\sqrt{2\pi}}e^{-\frac{y^2}{2\sigma^2}}dy + \mu\int_{-\infty}^\infty\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{y^2}{2\sigma^2}}dy
\end{aligned}$   
Let $g(y)=\frac{y}{\sigma\sqrt{2\pi}}e^{-\frac{y^2}{2\sigma^2}}$, then $g$ is an odd function, and hence it's   integral over $\mathbb{R}$ is $0$. Therefore,  
$\begin{aligned}
E[X]&=\mu\int_{-\infty}^\infty f_X(y)dy\\
&=\mu
\end{aligned}$  
And, 
  
$\begin{aligned}
Var(X)&=E[(X-E[X])^2]\\
&=E[(X-\mu)^2]\\
&=\int_{-\infty}^\infty\frac{(x-\mu)^2}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}dx\\
&=\int_{-\infty}^\infty\frac{y^2}{\sigma\sqrt{2\pi}}e^{-\frac{y^2}{2\sigma^2}}dy
\end{aligned}$  
Let $h(y)=\frac{y^2}{\sigma\sqrt{2\pi}}e^{-\frac{y^2}{2\sigma^2}}$, then $h$ is an even function, and hence,
$\begin{aligned}
Var(X)=2\int_0^\infty\frac{y^2}{\sigma\sqrt{2\pi}}e^{-\frac{y^2}{2\sigma^2}}dy
\end{aligned}$  
By substituting $t=\frac{y^2}{2\sigma^2}$.  $\therefore y^2=2\sigma^2t,dy=\frac{\sigma}{\sqrt{2t}}dt$. Therefore,  
$\begin{aligned}
Var(X)&=2\int_0^\infty\frac{2\sigma^2t}{\sigma\sqrt{2\pi}}e^{-t}\frac{\sigma}{\sqrt{2t}}dt\\
&=\frac{2\sigma^2}{\sqrt{\pi}}\int_0^\infty t^{\frac{1}{2}}e^{-t}dt\\
&=\frac{2\sigma^2}{\sqrt{\pi}}\Gamma\left(\frac{3}{2}\right)
\end{aligned}$  
Where $\Gamma$ is the Gamma function.  
We know that,  
$\Gamma(x+1)=x\Gamma(x)$ and $\Gamma\left(\frac{1}{2}\right)=\sqrt{\pi}$  
$\implies \Gamma\left(\frac{3}{2}\right)=\frac{1}{2}\Gamma\left(\frac{1}{2}\right)=\frac{\sqrt{\pi}}{2}$
 
$\begin{aligned}
Var(X)&=\frac{2\sigma^2}{\sqrt{\pi}}\times\frac{\sqrt{\pi}}{2}\\
&=\sigma^2
\end{aligned}$  
Hence, for a normal random variable $N(\mu,\sigma^2)$ the mean is $\mu$ and the variance is $\sigma^2$.

## Q6 : 
### (a)  For $\mu=0,\sigma=3$ :  
For a random variable $X$ with Normal distribution $N(\mu,\sigma^2)$,  
the PDF is $f_X(x)=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$,  
the CDF is   $F_X(x)=\phi\left(\frac{x-\mu}{\sigma}\right)$   
and the inverse of CDF is $F_X^{-1}(x)=\sigma\phi^{-1}(x)+\mu$   where $\phi$ denotes the CDF of a Standard Normal Distribution. 
  
  
**Inference** : The CDF inverse numbers form a gaussian distribution. 
The histogram also resembles the PDF.
  
The code for plot is-  

```python
x = np.sort(np.random.rand(10000))
y = norm.ppf(x,scale=3)
pdf_vals = norm.pdf(y,scale=3)

plt.hist(y,density=True,bins=40,label='histogram')
plt.plot(y,pdf_vals,lw=2,label='PDF')
plt.title('Gaussian')
plt.legend()
plt.show()
```

### (b)  Rayleigh density with $\sigma=1$  :  
  
the PDF is $f_X(x)=\frac{x}{\sigma^2}e^{-\frac{x}{2\sigma^2}}$,  
the CDF is $F_X(x)=1-e^{-\frac{x}{2\sigma^2}}$   
and the inverse of CDF is $F_X^{-1}(x)=\sigma\sqrt{-2ln(1-x)}$  

**Inference** : The CDF inverse values form a rayleigh distribution. 
The histogram resembles the PDF.  
  
The code for plot is-  

```python
from scipy.stats import rayleigh

x = np.sort(np.random.rand(10000))
y = rayleigh.ppf(x)
pdf_vals = rayleigh.pdf(y)

plt.hist(y,density=True,bins=40,label='histogram')
plt.plot(y,pdf_vals,lw=2,label='PDF')
plt.title('Rayleigh Distribution')
plt.legend()
plt.show()
```

### (c) Exponential density with $\lambda = 1.5$ : 


the PDF is $f_X(x)=\lambda e^{-\lambda x}$,  
the CDF is $F_X(x)=1-e^{-\lambda x}$   
and the inverse of CDF is $F_X^{-1}(x)=\frac{-ln(1-x)}{\lambda}$  
Also, the standard deviation $\sigma=\frac{1}{\lambda}$  
   
**inference** : The CDF inverse values form an exponential distribution. 
The histogram resembles the PDF.  

The code for plot is-  

```python
from scipy.stats import expon
x = np.sort(np.random.rand(10000))
y = expon.ppf(x,scale=1/1.5)
pdf_vals = expon.pdf(y,scale=1/1.5)

plt.hist(y,density=True,bins=40,label='histogram')
plt.plot(y,pdf_vals,lw=2,label='PDF')
plt.title('Plotting histogram and the PDF of Exponential Distribution')
plt.legend()
plt.show()
```

According to the **Cenral Limit Theorem**, if independent random variables are summed up, their properly normalized sum tends toward a normal distribution even if the original variables themselves are not normally distributed. This theorem can explain why our plot looks like this.
For $n$ independent random variables $X_1,X_2,...,X_n$ with mean $\mu$ and standard deviation $\sigma$,   let the resulting random variable be  
$\begin{aligned}
Z=\sum_{i=1}^nX_i
\end{aligned}$.  
Then $Z$ is approximately equal to $N(n\mu,n\sigma^2)$.  
Here, $n=500,\mu=\frac{1}{2},\sigma=\frac{1}{\sqrt{12}}$.
